```python
from langchain.chat_models import ChatOpenAI
import os

os.environ["OPENAI_API_KEY"] ="openai key"

llm = ChatOpenAI(model_name="gpt-4o-mini")
```

```python
from crewai_tools import WebsiteSearchTool

websearch_tool = WebsiteSearchTool()
```

```python
from crewai import Agent

web_scraper = Agent(
    name="Article Explorer",
    role="Artificial Intelligence Article Explorer",
    goal="Provide latest news of the AI industry for writing a LinkedIn post",
    backstory="An expert AI researcher specialized in gathering and summarizing the latest developments in artificial intelligence.",
    tools=[websearch_tool],
    verbose=True
)
```

```python
post_writer = Agent(
    role="Content Writer",
    goal="Analyze the article about the AI industry",
    backstory="A skilled LinkedIn post writer",
    tools=[],
    verbose=True
)
```

```python
from crewai import Task

web_scraper_task = Task(
    agent=web_scraper,
    name="Web Scraper Task",
    description="Search the web for the latest AI trends and provide a summary.",
    expected_output="Provide the top 5 most trending and recent news articles about Artificial Intelligence."
)
```

```python
write_post_task = Task(
    agent=post_writer,
    name="LinkedIn Post Writing Task",
    description="Analyze the AI news summary and craft an engaging LinkedIn post that highlights the key trends and insights.",
    expected_output="A polished LinkedIn post of around 150-300 words, written in a professional and conversational tone.",
    output_file="linkedin_post.txt"
)
```

```python
from crewai import Crew
crew = Crew(
    agents=[web_scraper, post_writer],
    tasks=[web_scraper_task, write_post_task],
    verbose=True,
    planning=True

)

crew.kickoff()
```
